```python
import pandas as pd
```

```python
df=pd.read_csv("Hospital_patients_datasets.csv")
df
```

```python
duplicates = df.duplicated().sum()
duplicates
```

```python
null_values = df.isnull().sum()
null_values
```

```python
# Convert columns to datetime and then to date-only
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay']).dt.date
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay']).dt.date
```

```python
df
```

```python
df.rename({'Hipertension': 'Hypertension', 'Handcap' : 'Handicap', 'SMS_received' : 'SMSReceived', 'No-show' : 'NoShow'}, axis=1, inplace=True)
df
```

```python
df.drop(['PatientId', 'AppointmentID', 'Neighbourhood'], axis=1, inplace=True)
df
```

```python

# Remove rows where 'Age' equals 0
df = df[df['Age'] != 0]

# Define age group bins and labels
bins = [1, 20, 40, 60, 80, 100, 130]
labels = ['1-20', '21-40', '41-60', '61-80', '81-100', '101-130']

# Create 'Age_group' column
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True, include_lowest=True)

# Display the updated DataFrame
print(df.head())
```

```python
# Remove rows where 'Age' equals 0
df = df[df['Age'] != 0].copy()

# Define age group bins and labels
bins = [1, 20, 40, 60, 80, 100, 130]
labels = ['1-20', '21-40', '41-60', '61-80', '81-100', '101-130']

# Create 'Age_group' column using .loc
df.loc[:, 'Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True, include_lowest=True)

# Display the updated DataFrame
print(df.head())
```

```python
df.drop(["Age"], axis=1, inplace=True)
df
```

```python
import numpy as np
df['NoShow'] = np.where(df['NoShow'] == 'Yes', 1, 0)
df
```

```python
df.to_csv("patients.csv", index=False)
```

```python
sql = """SELECT AppointmentDay, COUNT(*) AS appointment_count
FROM patients
GROUP BY AppointmentDay
ORDER BY AppointmentDay

;
"""
df = pd.read_sql_query(sql, con = engine.connect())
df
```

```python
sql = """SELECT AVG(appointment_count) AS avg_appointments_per_day
FROM (
    SELECT AppointmentDay, COUNT(*) AS appointment_count
    FROM patients
    GROUP BY AppointmentDay
) daily_counts
;
"""
df = pd.read_sql_query(sql, con = engine.connect())
df
```

```python
sql = """SELECT AppointmentDay, COUNT(*) AS appointment_count
FROM patients
GROUP BY AppointmentDay
ORDER BY appointment_count DESC
LIMIT 1;
"""
df = pd.read_sql_query(sql, con = engine.connect())
df
```
